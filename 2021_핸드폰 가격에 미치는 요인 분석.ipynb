```python
import csv
f = open('train.csv')
data = csv.reader(f)
next(data)
data = list(data)

print(len(data))
```

```python
for row in data:
    for i in range(21):
        if i ==2 or i == 7 :
            row[i] = float(row[i])
        else :
            row[i] = int(row[i])
print(data[0])
```

```python
price_range = [0 for i in range(4)]
for row in data:
    for i in range(4):
        if row[-1] == i:
            price_range[i] += 1
print(price_range)
    
```

```python
battery_power = [0 for i in range(4)]
int_memory = [0 for i in range(4)]
mobile_wt = [0 for i in range(4)]
n_cores = [0 for i in range(4)]
pc = [0 for i in range(4)]
ram = [0 for i in range(4)]

for row in data:
    for i in range(4):
        if row[-1] == i:
            battery_power[i] += row[0]
            int_memory[i] += row[6]
            mobile_wt[i] += row[8]
            n_cores[i] += row[9]
            pc[i] += row[10]
            ram[i] += row[13]
            
print(battery_power,int_memory,mobile_wt,n_cores,pc,ram)


battery_power = [i/500 for i in battery_power]
int_memory = [i/500 for i in int_memory]
mobile_wt = [i/500 for i in mobile_wt]
n_cores = [i/500 for i in n_cores]
pc = [i/500 for i in pc]
ram = [i/500 for i in ram]


print(battery_power,int_memory,mobile_wt,n_cores,pc,ram)
```

```python
import matplotlib.pyplot as plt
import numpy

price = ['low cost' , 'medium cost', 'high cost', 'very high cost']
plt.figure(dpi = 150)
plt.rc('font', family='Malgun Gothic')
plt.title('가격대별 휴대폰 성능 비교')

x = numpy.arange(len(price))

plt.bar(x-0.0, battery_power, color='red',width = 0.1, label = '배터리')
plt.bar(x+0.1, ram, color='springgreen',width = 0.1, label = '램')
plt.bar(x+0.2, int_memory, color='lavender',width = 0.1,label = '내장 메모리')
plt.bar(x+0.3, mobile_wt, color='brown',width = 0.1, label = '휴대폰 무게')
plt.bar(x+0.4, n_cores, color='cyan',width = 0.1,label = '중앙처리장치 개수')
plt.bar(x+0.5, pc, color='orange',width = 0.1, label = '카메라 픽셀')

plt.xticks(x, price) 

plt.xlabel('가격산정도') 
plt.ylabel('평균치')

plt.legend()
plt.show()
```
